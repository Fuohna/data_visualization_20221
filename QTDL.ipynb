```python
import dash
from jupyter_dash import JupyterDash
from dash import dcc, html, Dash
import pandas as pd
import numpy as np
import plotly.express as px
import pycountry_convert as pc
```

### Import, preprocessing data

```python
netflix = pd.read_csv("/content/drive/MyDrive/cần in/HUST_SoICT_20221/Quản trị dữ liệu/netflix_titles.csv")
netflix.shape
# netflix.head()
netflix.isnull().sum()
```

```python
director_df = netflix.copy()
```

```python
netflix.info()
```

```python
netflix.drop(['show_id', 'director'], axis=1, inplace=True)
```

```python
netflix['date_added'] = pd.to_datetime(netflix['date_added'])

netflix.dropna(inplace=True)
netflix = netflix.reset_index(drop=True)
netflix.shape
netflix.info()
```

```python
netflix.head()
```

```python
def country_to_continent_alpha3(country_name):
    country_alpha3 = pc.country_name_to_country_alpha3(country_name)
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_alpha3, country_continent_name
```

```python
netflix['country'] = netflix['country'].apply(lambda x: x.split(',')[0])
netflix = netflix[netflix['country'] != "Soviet Union"]
countries = netflix['country'].unique()

country_alpha3 = []
country_continent = []

for country_name in netflix['country']:
  alpha3, continent = country_to_continent_alpha3(country_name)
  country_alpha3.append(alpha3)
  country_continent.append(continent)

netflix['iso_alpha3'] = country_alpha3
netflix['continent'] = country_continent
# netflix
```

```python
country = netflix['country'].unique()
ratings = netflix['rating'].unique()
# print(country)
```

## Visualization 

### Những nội dung trên Netflix được sản xuất bởi quốc gia nào?

```python
map_df = pd.DataFrame(columns=["type", "country", "iso_alpha3", "continent", "year"])
map_df["type"] = netflix["type"]
map_df["country"] = netflix["country"]
map_df["iso_alpha3"] = netflix["iso_alpha3"]
map_df["continent"] = netflix["continent"]
map_df["year"] = netflix["date_added"].dt.year

def cal_size(country):
  count = 0
  percent = 0
  for ct_name in map_df["country"]:
    if ct_name == country:
      count+=1
    percent = float(count/len(map_df))
  return percent

percent_of_show = []
for ct_name in map_df["country"]:
  percent_of_show.append(cal_size(ct_name))

map_df["%_of_total"] = percent_of_show
```

```python
fig_buble_map = px.scatter_geo(map_df, 
                              title="Distribution of Content Production by Country", 
                              locations="iso_alpha3", 
                              color="continent", 
                              size="%_of_total", 
                              size_max=30,
                              hover_name="country",
                              animation_frame="year",
                              projection="natural earth")
fig_buble_map.update_layout(
    # paper_bgcolor='#F5F5F1',
    # plot_bgcolor='#F5F5F1',
    title_font=dict(size=20, color='#221f1f'),
    font=dict(color='#221f1f'), 
)
fig_buble_map.show()
```

### Chương trình truyền hình hoặc bộ phim nào có xếp hạng cao nhất? 

```python
movie_tvshow_df = netflix.copy()

# making 2 df one for tv show and another for movie with rating 
tvshow_df = movie_tvshow_df[movie_tvshow_df['type']=='TV Show'][['rating', 'type']].rename(columns={'type':'tv_show'})
movie_df = movie_tvshow_df[movie_tvshow_df['type']=='Movie'][['rating', 'type']].rename(columns={'type':'movie'})
tvshow_df = pd.DataFrame(tvshow_df.rating.value_counts()).reset_index().rename(columns={'index':'tv_show'})
tvshow_df['rating_final'] = tvshow_df['rating'] 
tvshow_df['rating'] *= -1
movie_df = pd.DataFrame(movie_df.rating.value_counts()).reset_index().rename(columns={'index':'movie'})
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig_movie_tvshow = make_subplots(rows=1, cols=2, specs=[[{}, {}]], shared_yaxes=True, horizontal_spacing=0)

fig_movie_tvshow.add_trace(go.Bar(x=tvshow_df['rating'], y=tvshow_df['tv_show'], orientation='h', showlegend=True, 
                        text=tvshow_df.rating_final, name='TV Show', marker_color='#221f1f'), 1, 1)
fig_movie_tvshow.add_trace(go.Bar(x=movie_df['rating'], y=movie_df['movie'], orientation='h', showlegend=True, 
                        text=movie_df.rating, name='Movie', marker_color='#b20710'), 1, 2)

fig_movie_tvshow.update_layout(
    margin=dict(t=80, b=0, l=70, r=40),
    hovermode="y unified", 
    xaxis_title=" ", 
    yaxis_title=" ", 
    # height=400,
    # width=400,
    plot_bgcolor='#FFF', 
    paper_bgcolor='#FFF', 
    title_font=dict(size=20, color='#221f1f'),
    font=dict(color='#221f1f'),
    legend=dict(orientation="h", yanchor="bottom", y=1, xanchor="center", x=0.5), 
    hoverlabel=dict(bgcolor="white", font_size=13)
)
fig_movie_tvshow.show()
```

### Thời gian nào tốt nhất để phát hành nội dung?

```python
month_df = netflix.copy()
month_df['date_added'] = month_df['date_added'].dt.month
month_df = month_df['date_added'].value_counts().reset_index().rename(columns={'index':'month','date_added':'count'})
month_df['month_name'] = month_df['month'].replace({1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'June', 7:'July', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'})
# month_df
```

```python
fig_month = px.funnel(month_df, x='count', y='month_name', title='Best month for releasing Content', color_discrete_sequence=['#b20710'])
fig_month.update_xaxes(showgrid=False, ticksuffix=' ', showline=True)
fig_month.update_traces(hovertemplate=None, marker=dict(line=dict(width=0)))
fig_month.update_layout(
    margin=dict(t=80, b=0, l=70, r=40),
    hovermode="y unified", 
    xaxis_title=" ", 
    yaxis_title=" ", 
    # height=400,
    width=800,
    plot_bgcolor='#FFF', 
    paper_bgcolor='#FFF', 
    title_font=dict(size=20, color='#221f1f'),
    font=dict(color='#221f1f'),
    legend=dict(orientation="h", yanchor="bottom", y=1, xanchor="center", x=0.5), 
    hoverlabel=dict(bgcolor="white", font_size=13)
)
```

### Sự phát triển của các nội dung được sản xuất qua các năm 

```python
df1=netflix[['type','release_year']]
df1=df1.rename(columns={"release_year": "Release Year"})
df2=df1.groupby(['Release Year','type']).size().reset_index(name='Total Content')
df2=df2[df2['Release Year']>=2010]
fig_line_type = px.line(df2, x="Release Year", y="Total Content", color='type', color_discrete_sequence=['#b20710', '#221f1f'])
fig_line_type.update_layout(
    title='Content added over the years',
    margin=dict(t=80, b=0, l=70, r=40),
    hovermode="y unified", 
    xaxis_title=" ", 
    yaxis_title=" ",
    width = 500,
    height = 400,
    plot_bgcolor='#FFF', 
    paper_bgcolor='#FFF', 
    title_font=dict(size=20, color='#221f1f'),
    font=dict(color='#221f1f'), 
    legend=dict(orientation="h", yanchor="bottom", y=1, xanchor="center", x=0.5), 
    hoverlabel=dict(bgcolor="white", font_size=13)
)
fig_line_type.show()
```

### Phân phối các thể loại được xem nhiều trên Netflix 

```python
def getlist(data):
  data_list =  list()
  for v in data:
    values = v.split(", ")
    for i  in values:
      data_list.append(i)
  return list(data_list)
genre_list =  getlist(netflix['listed_in'])
df_genre = pd.DataFrame(pd.value_counts(genre_list).reset_index().rename(columns={'index':'genre', 0:'count'}))
```

```python
fig_tree = px.treemap(df_genre, path=[px.Constant("Distribution of Genres"), 'count','genre'], color_continuous_scale='RdBu', color = 'count')
fig_tree.update_layout(
    title='Watched Genres on Netflix',
    margin=dict(t=80, b=0, l=70, r=40),
    hovermode="y unified", 
    plot_bgcolor='#FFF', 
    paper_bgcolor='#FFF', 
    title_font=dict(size=20, color='#221f1f'),
    font=dict(color='#221f1f'), 
    legend=dict(orientation="h", yanchor="bottom", y=1, xanchor="center", x=0.5), 
    hoverlabel=dict(bgcolor="white", font_size=13),
)
```
